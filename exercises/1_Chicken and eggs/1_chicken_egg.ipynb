```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,6,9,10,15,13]).reshape(-1,1)
y = np.array([2,6,8,11,15,21,30,40,34]).reshape(-1,1)
```

```python
def plot_data():
    plt.scatter(X,y,label='Data');
    plt.xlabel('Chicken')
    plt.ylabel('Egg')
```

```python
def plot_regression_line(label):
    a = [X.min() ,X.max()]
    b = [np.array([1, X.min()]).dot(w), np.array([1, X.max()]).dot(w)]
    plt.plot(a, b, '--', label=label);
```

# Normal equsion

```python
X1 = np.c_[np.ones(len(X)), X]
w = np.linalg.inv(X1.T.dot(X1)).dot(X1.T).dot(y)

plot_data()
plot_regression_line('Reg. Line')
plt.legend();
plt.show()
```

# Gradient decent with Matrix operations

```python
cost = []
m = len(X1)
alpha = 0.025
w = np.zeros((X1.shape[1],1))
for epoch in range(0,20):
    plot_regression_line(label=str(epoch))
    y_hat = X1.dot(w)
    J = ((y_hat-y)**2).sum()/(2*m)
    cost.append(J)
    
    grad = X1.T.dot(y_hat-y)
    w -= alpha*grad/m
 
plot_regression_line(label=str(epoch))
plot_data()
plt.legend()
plt.show()

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(cost);
    
cost[-1]  
```
