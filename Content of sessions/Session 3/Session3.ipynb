```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [22.0, 21.0, 19.0, 16.0, 15.0, 13.0, 11.0, 7.0, 4.0, 3.0]
y = [112.4, 119.7, 131.3, 168.2, 179.0, 202.6, 229.3, 293.6, 373.2, 405.7]
```

```python
def plot_data():
    plt.scatter(X,y, label='Data')
    plt.xlabel('Age')
    plt.ylabel('Price')
```

```python
def plot_regression_line(label):
    
    xmin = sorted(X)[0]
    xmax = sorted(X)[-1]
    xx =[xmin, xmax]
    yy =[w*xmin+b, w*xmax+b]
    plt.plot(xx,yy,'--',label=label)
```

```python
def predict(x):
    return x*w+b
```

```python
m = len(X)
alpha = 0.009
w = 0
b = 0
cost = []
for epoch in range(3000):
    if epoch%300==0:
        plot_regression_line(str(epoch))
    J = 0
    g_w = 0
    g_b = 0
    for i in range(m):
        y_hat = w*X[i]+b
        J += (y_hat-y[i]) ** 2
        
        g_w += (y_hat - y[i]) * X[i]
        g_b += (y_hat - y[i])
        
    J = J/(2*m)
    #print(J)
    cost.append(J)
    g_w = g_w/m
    g_b = g_b/m
    
    w -= alpha * g_w
    b -= alpha * g_b

plot_regression_line(str(epoch))
plot_data()
plt.legend()
plt.show()
```

```python
print(cost[-1])
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.plot(cost)
```
