# Classification

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([ [1,2] , [1.5,2], [0.5,0.5] , [1.5,1.5] , [3.5,3.7] , [4,4] , [3,3.5] , [4,3] , [8,4]])
y = np.array( [0,0,0,0,1,1,1,1,1] )

X = np.c_[np.ones(len(X)) , X]
```

```python
X
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
alpha = 0.01
w = np.zeros(len(X[0]))
m = len(X)
cost = []

for epoch in range(100000):
    y_hat = sigmoid(X.dot(w))
    grad = (1/m) * X.T.dot(y_hat - y)
    w -= alpha * grad

    J = (-1/m) * ( y*np.log(y_hat) + (1-y)*np.log(1-y_hat) ).sum()
    cost.append(J)

plt.plot(cost)
print(f'Error: {J}')
```
