Libraries:
- NumPy, a popular library for scientific computing
- Matplotlib, a popular library for plotting data
- scikit-learn, Simple and efficient tools for predictive data analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random
import copy
import math
```

[sklearn](https://scikit-learn.org/)

```python
from sklearn import linear_model
```

* data

```python
xx = np.array([-1,1,2,3,4,5,6])
y = np.array([2,3,4,6,6,6,8.5])

X = [[i] for i in xx]

print(f'xx: {xx}')
print(f'X: {X}')
print(f'y: {y}')
```

* plot

```python
plt.scatter(xx,y)
```

* model
    * model = model     creat model
    * model.fit(x,y)    train the model
    * model.predict(new_x)
  

```python
reg = linear_model.LinearRegression()
```

* fit (train)

```python
reg.fit(X,y)
```

* predict

```python
x_new = [3]
y_predict = reg.predict([x_new])
print(y_predict)
```

* plot

```python
plt.scatter(xx, y, s = 50)
plt.scatter(x_new, y_predict)
```

```python
plt.scatter(xx,y,s=50)
x_p = [[min(xx)-1], [max(xx)+1]]
y_p = reg.predict(x_p)
plt.plot(x_p, y_p, c = 'red')
plt.scatter(x_new, y_predict, marker = '*', s=300)
plt.title("Housing Prices")
plt.ylabel('Price')
plt.xlabel('Size')
plt.show()
```
